#Importando o Pandas
Biblioteca responsável pela manipulação de datasets

```python
import pandas as pd
```

#importando o dataset

```python
df = pd.read_csv("banco_clientes.csv")
df
```

#Padronizando os Nomes

```python
df['nome'] = df['nome'].str.upper()
df
```

#Padronizando os telefones

```python
df['telefone'] = df['telefone'].str.replace(')', '',regex =  False).str.replace('(', '',regex =  False)
df
```

#Padronizando os endereços

```python
df['CEP'] = df['endereco'].str.extract('(\d{5}-\d{3})')
df
```

##Precisamos remover os Ceps do endereço

```python
df['endereco'] = df['endereco'].str.replace('\d{5}-\d{3}', '', regex = True).str.strip() #o strip remove o espaço em branco.
df
```

##Separando o n° da rua do nome da rua

```python
df[["rua","numero"]] = df.endereco.str.replace(",","").str.split('\s(?=\d+$)',regex=True, expand=True)
df
```

#Removendo Endereço
Já que temos o cep, rua e número separados podemos remover a coluna endereço

```python
df = df.drop(["endereco"], axis=1)
df
```
